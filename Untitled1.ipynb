```python
import pandas as pd
```

```python
data = pd.read_excel('data.xlsx', sheet_name = None)
txns = data['Transactions']
persons = data['Persons']
items = data['Items']
```

```python
txns = txns.fillna(pd.NA)
txns['person_id'] = txns.person_id.astype(pd.Int64Dtype())
txns.head()
```

```python
persons = persons.fillna(pd.NA)
persons['phone'] = '0' + persons.phone.astype(pd.Int64Dtype()).astype(pd.StringDtype())
persons.head()
```

```python
result = txns[txns.category == 'sales'].merge(
    txns[txns.category == 'purchases'],
    on = 'item_id',
    suffixes = ['_sales', '_purchases'],
    how = 'left'
)
result = result.groupby('date_sales').agg(
    sale = ('amount_sales', 'sum'),
    qty_sold = ('item_id', 'count'),
    cost = ('amount_purchases', 'sum'),
)
result['profit'] = result.sale - result.cost
result
```
