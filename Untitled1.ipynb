```python
import pandas as pd
```

```python
data = pd.read_excel('data.xlsx', sheet_name = None)
txns = data['Transactions']
persons = data['Persons']
items = data['Items']
```

```python
txns.head()
```

```python
txns = txns.fillna(pd.NA)
txns['person_id'] = txns.person_id.astype(pd.Int64Dtype())
txns.head()
```

```python
persons.head()
```

```python
persons = persons.fillna(pd.NA)
persons['phone'] = '0' + persons.phone.astype(pd.Int64Dtype()).astype(pd.StringDtype())
persons.head()
```

```python
items.head()
```

```python
items2 = items.category.drop_duplicates().reset_index(drop = True, name = 'category').to_frame()
items2['item_id'] = items2.index + 1
items2 = items2[['item_id', 'category']]
items2
```

```python
txns.head()
```

```python
txns2 = txns.copy()
txns2['item_id'] = txns2.item_id.map(items.set_index('item_id').category).map(items2.set_index('category').item_id).fillna(pd.NA).astype(pd.Int64Dtype())
txns2['qty'] = pd.NA
txns2.tail()
```

```python
txns2['item_id'] = txns2.item_id.map(items2.set_index('item_id').category)
txns2['item_id'] = txns2.item_id.fillna('Unknown')
txns2.loc[txns2.category == 'sales', :] = txns2[txns2.category == 'sales'].groupby(['date',  'category', 'type', 'person_id', 'item_id']).agg(
    amount = ('amount', 'sum'),
    qty = ('item_id', 'count')
).reset_index()
txns2
```

```python

purchases2 = txns2[txns2.category == 'purchases'].groupby(['date',  'category', 'type', 'item_id']).agg(
    amount = ('amount', 'sum'),
    qty = ('item_id', 'count')
).reset_index()
purchases2['person_id'] = pd.NA
purchases2
```

```python
txns2 = pd.concat([txns2[~(txns2.category == 'purchases')], purchases2], axis = 0)
```

```python
txns2 = txns2.dropna(how = 'all').reset_index(drop = True)
txns2.tail()
```

```python
txns2
```

```python
txns2 = txns2.rename({'item_id': 'item_category'}, axis = 1)
txns2.head()
```

```python
txns2['txn_id'] = txns2.index + 1
txns2.head()
```

```python
txns2['date'] = txns2.date.dt.date
persons['cus_date'] = persons.cus_date.dt.date

with pd.ExcelWriter('data2.xlsx') as writer:
    txns2.to_excel(writer, sheet_name = 'Transactions', index = False)
    persons.to_excel(writer, sheet_name ='Persons', index = False)
```
