```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('data.xlsx', sheet_name = None)
data.keys()
```

```python
items = data['Items']
```

```python
items.head()
```

```python
sales = data['Sales']
sales.head()
```

```python
pnl = sales.merge(items, on = 'item_id', how = 'left')
pnl = pnl.rename({
    'date_x': 'sell_date',
    'date_y': 'buy_date'
}, axis = 1)
pnl.head()
```

## Daily Profit

```python
daily_profit = (pnl.sell_price - pnl.cost_price).groupby(pnl.sell_date).sum().reset_index(name = 'profit')
daily_profit
```
