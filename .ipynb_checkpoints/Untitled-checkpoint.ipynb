```python
import pandas as pd
```

```python
data = pd.read_excel('data2.xlsx', sheet_name = None)
items = data['Items']
txns = data['Transactions']
persons = data['Persons']
```

```python
is_sale = txns.category == 'sales'
is_purchase = txns.category == 'purchases'
```

```python
txns[is_sale][~txns[is_sale].item_id.isin(txns[is_purchase].item_id)]
```

```python
txns[is_purchase][txns[is_purchase].item_id.isin(txns[is_sale].item_id)].amount.sum()
```

```python
txns[is_sale][txns[is_sale].item_id.isin(txns[is_purchase].item_id)].amount.sum()
```

```python
txns[is_purchase].amount.sum()
```

```python
result = txns[is_sale].groupby('date').amount.sum().reset_index(name = 'value')
result
```

```python
result = txns[is_sale].merge(items, on = 'item_id', how = 'left', suffixes = ['_txns', '_items'])
result = result.groupby('category_items').agg(
    sales = ('amount', 'sum'),
    qty_sold = ('item_id', 'count')
).sort_values(by = ['sales', 'qty_sold'], ascending = False).reset_index()
    
result
```

```python
inv = items[~items.item_id.isin(txns[is_sale].item_id)]
inv = inv.merge(txns[is_purchase][['item_id', 'amount', 'date']], on = 'item_id', how = 'left')
inv = inv.rename({
    'amount': 'cost',
    'date': 'purchase_date'
}, axis = 1)

inv
```
