```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('data.xlsx', sheet_name = None)
data.keys()
```

```python
cus = data['Customers'].dropna()
cus.head()
```

```python
exp = data['Expenses']
exp.head()
```

```python
items = data['Items']
```

```python
items.head()
```

```python
sales = data['Sales']
sales.head()
```

```python
pnl = sales.merge(items, on = 'item_id', how = 'left')
pnl = pnl.rename({
    'date_x': 'sell_date',
    'date_y': 'buy_date'
}, axis = 1)
pnl.head()
```

## Daily Profit

```python
pnl['profit'] = (pnl.sell_price - pnl.cost_price)
pnl.groupby('sell_date').agg(
    profit = ('profit', 'sum'),
    sales = ('sell_price', 'sum'),
    items_sold = ('item_id', 'count')
).reset_index().rename({'sell_date': 'date'}, axis = 1)
```

## Inventory state

```python
items[~items.item_id.isin(sales.item_id)].groupby('category').item_id.size().sort_values().reset_index(name = 'available_items')
```

## Customer Analysis

```python
cus_sales = cus.merge(sales, on = 'customer_id', how = 'left')
cus_sales.head()
```

```python
cus_value = cus_sales.groupby('customer_id').agg(
    value = ('sell_price', 'sum'),
    purchases = ('item_id', 'count'),
    mean_value = ('sell_price', 'mean')
).sort_values(by = ['value', 'purchases'], ascending = False).reset_index()

cus_value['name'] = cus_value.customer_id.map(cus.set_index('customer_id').name)
cus_value
```

```python
addr_value = cus_sales.groupby('addr').agg(
    value = ('sell_price', 'sum'),
    purchases = ('item_id', 'count'),
    mean_value = ('sell_price', 'mean')
).sort_values(by = 'value', ascending = False).reset_index()

addr_value
```

## Best selling items

```python
(sales.groupby(sales.item_id.map(items.set_index('item_id')['category']))
 .size().sort_values(ascending = False).reset_index(name = 'num_sold')
 .rename({'item_id': 'category'}, axis = 1))
```
